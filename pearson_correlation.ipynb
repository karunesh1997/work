First we import the packages that we need:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

Letâ€™s read the data and put it in a dataframe.

```python
data_test = pd.read_csv('heart_failure.csv')
data_test.head()
```

# Calculating the Pearson Corellation

```python
data_corr = data_test.corr(method='pearson')
data_corr
```

To make this look beautiful and easier to interpret, add this after calculating the Pearson coefficient of correlation.

```python
sb.heatmap(data_corr, 
            xticklabels=data_corr.columns,
            yticklabels=data_corr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```
